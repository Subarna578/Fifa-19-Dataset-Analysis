```python
DF['total_score']=(DF.Overall/10+DF.Potential/10+DF['International Reputation']+DF['Skill Moves'])*3.333
```

```python
DF.total_score
```

```python
DF.sample(3)
```

```python
DF.Nationality
```

```python
DF=DF[DF.Age<=30]
```

```python
DF.sample()
```

```python
best=DF.groupby('Age')['total_score'].agg(np.max).reset_index()
```

```python
best
```

```python
age_scores=list(zip(best.Age,best.total_score))
```

```python
age_scores
```

```python
best_players=pd.DataFrame(columns=DF.columns)
```

```python
best_players
```

```python
for index,i in enumerate(age_scores):
    best_players=best_players.append(DF[(DF.Age==i[0]) & (DF.total_score==i[1])],ignore_index=True)
```

```python
best_players
```

```python
#here we can see some of the greatest players who are under or equal to 30 years of age.
#For eg. Neymar Jr, G.Bale, Marcelo, P.Dybala
```
